```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Loading the dataset
df = pd.read_csv('USvideos.csv')  
# Calculate the number of numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
num_columns = len(numeric_columns)

# Determine grid size based on the number of columns
cols = 2  # Number of columns in the grid
rows = math.ceil(num_columns / cols)  # Calculate rows required

# Violin plots to identify outliers and distributions
plt.figure(figsize=(12, 6))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    sns.violinplot(x=df[column])
    plt.title(f'Violin Plot of {column}')
plt.tight_layout()
plt.show()

# KDE plots for the distribution of variables
plt.figure(figsize=(12, 6))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    sns.kdeplot(df[column], shade=True)
    plt.title(f'KDE Plot of {column}')
plt.tight_layout()
plt.show()
```
